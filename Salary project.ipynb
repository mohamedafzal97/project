# Transforming data to remove skewness

```python
#independent column
x=df1.iloc[:,:-1].values
#dependent column or target column
y=df1.iloc[:,-1].values
```

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
```

```python
#checking skewness after removing skewness
pd.DataFrame(x).skew()
```

```python
#scaling independent data using StandardScaler
#mean=0, std=1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

# splitting data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20,random_state=32)
```

```python
#checking the shape
x_train.shape, y_train.shape, x_test.shape , y_test.shape
```

# Since our target column is continous data we will use RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
#creating instance for RandomForestRegressor
RFR = RandomForestRegressor(n_estimators = 150, random_state = 32)
```

```python
# training model using training data
RFR.fit(x_train, y_train)
```

```python
#predicting the data
y_pred=RFR.predict(x_test)
```

# Evaluating:

```python
#dataframe of actual data and predicted data
diff=y_test-y_pred
data={'Actual':y_test,'Predicted':y_pred,'Difference':diff}
pd.DataFrame(data)
```

```python
# calculating the mean absolute error
abserror= abs(y_pred - y_test)
print('mean absolute error:', np.mean(abserror))
```

```python
#calculating the percentage of mean absolute rror
mean_abs_percentage_error= 100 * (abserror / y_test)
print('mean absolute percentage error:', round(np.mean(mean_abs_percentage_error),2),'%')
```

```python
#calculation accuracy of model
accuracy = 100 - np.mean(mean_abs_percentage_error)
print('Accuracy:', round(accuracy, 2), '%.')
```
