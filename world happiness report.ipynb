# Data preprocessing

# Data cleaning:

```python
#checking correlation between target column(Happiness Rank) and other columns
df.corr()['Happiness Score']
```

```python
# droping the colmns which has more outliers
df1= df.drop(['Standard Error','Trust (Government Corruption)'], axis =1)
```

```python
df1
```

# Removing outliers:

using Zscore technique

```python
from scipy.stats import zscore
```

```python
z=np.abs(zscore(df1.iloc[:,3:]))
z.shape
```

```python
threshold = 3   # rule of thumb
#index number
print(np.where(z>3))
```

```python
len(np.where(z>3)[0])
```

```python
df_new=df1[(z<3).all(axis=1)]
print('old dataframe',df1.shape)
print('new dataframe',df_new.shape)
print('no of rows dropped',df1.shape[0]-df_new.shape[0])
```

# percentage data loss:

```python
loss_percentage = (158-154)/158*100
print(loss_percentage,'%')
```

# Transforming data to remove skewness

```python
df_new
```

```python
#independent column
x=df_new.iloc[:,4:]
#dependent column or target column
y=df_new.iloc[:,3]
```

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
```

```python
#checking skewness after removing skewness
pd.DataFrame(x).skew()
```

# since our target column is continous data we will use LinerRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
#splitting the data to train and to test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=42)
```

```python
#shape of data to be trained
x_train.shape
```

```python
#shape of data to be trained 
y_train.shape
```

```python
#shape of data which is tested 
x_test.shape 
```

```python
#shape of actual output
y_test.shape
```

```python
#creating instance for linearRegression
lm=LinearRegression()
```

```python
#training the Model
lm.fit(x_train,y_train)
```

```python
#checking coeffient
lm.coef_
```

```python
#checking intercept
lm.intercept_
```

```python
#score of trained data
lm.score(x_train,y_train)
```

```python
#predicting  the target value
y_pred=lm.predict(x_test)
```

```python
print("predict happiness score:",y_pred)
print("actual happiness score:",y_test)
```

```python
print('error: ')
print('mean absolue error: ',mean_absolute_error(y_test,y_pred))
print('mean squared error: ',mean_squared_error(y_test,y_pred))

#calculating the error
print('\nroot mean squared error: ',np.sqrt(mean_squared_error(y_test,y_pred)))
```

# Conclusion:

From the above outcome we can see that our model has almost 0 root mean squared error i.e. our model is working almost 100 percent.
