# Model Building and Evaluation

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

# Since our target variable is continous data we use Regression Models

# Linear Regression

### Finding Best Random State for LinerRegression

```python
from sklearn.linear_model import LinearRegression

maxAcc=0
maxRS=0

for i in range(1,1500):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20,random_state=i)
    #creating instance for linearRegression
    lm=LinearRegression()
    #training the Model
    lm.fit(x_train,y_train)
    #predicting  the target test value
    pred_lm=lm.predict(x_test)
    acc=r2_score(y_test,pred_lm)
    
    if acc>maxAcc:
        maxAcc=acc
        maxRS=i
        print('Accuracy', maxAcc, 'random_state', maxRS)
```

from the above we can see that the best random state is 47 which is giving 0.98 accuracy.

# LinearRegression with best random state

```python
x_train1,x_test1,y_train1,y_test1=train_test_split(x_scaled,y,test_size=0.20,random_state=47)


#creating instance for linearRegression
lm=LinearRegression()
#training the Model
lm.fit(x_train1,y_train1)
#predicting  the target value
lm_pred=lm.predict(x_test1)
```

# Evaluation

```python
# calculating the mean absolute error
from sklearn.metrics import mean_absolute_error

abserror=mean_absolute_error(y_test1,lm_pred)

#calculating the percentage of mean absolute rror
mean_abs_percentage_error= 100 * (abserror / y_test1)

print(' mean absolute error:', abserror,'\n','mean absolute error percentage:', round(np.mean(mean_abs_percentage_error),2),'%')
```

```python
#calculating mean squared error
from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y_test1,lm_pred)

#calculating the percentage of mean squared error
mse_percentage= 100 * (mse / y_test1)

print(' mean squared error:', mse,'\n','mean squared error percentage:', round(np.mean(mse_percentage),2),'%')
```

```python
#calculating the mean squared log error
from sklearn.metrics import mean_squared_log_error

msle=mean_squared_log_error(y_test1,lm_pred)
print(' mean squared log error:', msle)
```

```python
#calculating r2_score
from sklearn.metrics import r2_score

print('r2_score:', r2_score(y_test1,lm_pred),'\n\nAccuracy:',round(100*(r2_score(y_test1,lm_pred)),2),'%')
```

# cross-validation of the model

```python
from sklearn.model_selection import cross_val_score

test_score=r2_score(y_test1,lm_pred)

#Linear Regression
scr=cross_val_score(lm,x,y,cv=4)
print('cross validation score of LinearRegression at cross fold 4: ',100*scr.mean(), 'and accuracy score of testing model is:',test_score*100   )
```

# Random Forest Regressor

# Finding Best Random State for Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

maxAcc=0
maxRS=0

for i in range(1,1500):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20,random_state=i)
    #creating instance for RandomForestRegressor
    RFR = RandomForestRegressor(n_estimators = 20, random_state = 40)
    # training model using training data
    RFR.fit(x_train, y_train)
    #predicting the data
    pred_RFR=RFR.predict(x_test)
    acc=r2_score(y_test,pred_RFR)
    
    if acc>maxAcc:
        maxAcc=acc
        maxRS=i
        print('Accuracy', maxAcc, 'random_state', maxRS)
```

from the above we can see that the best random state is 1444 which is giving 0.92 accuracy.

# Random Forest Regressor with best random state

```python
x_train2,x_test2,y_train2,y_test2=train_test_split(x_scaled,y,test_size=0.20,random_state=1444)


#creating instance for RandomForestRegressor
RFR = RandomForestRegressor(n_estimators = 20, random_state = 40)

# training model using training data
RFR.fit(x_train2, y_train2)

#predicting the data
RFR_pred=RFR.predict(x_test2)
```

# Regularization

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
```

```python
maxAcc=0
maxRS=0

for i in range(1,1500):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20,random_state=i)
    ls=Lasso(alpha=1,random_state=0)
    ls.fit(x_train,y_train)
    ls_pred=ls.predict(x_test)
    acc=r2_score(y_test,ls_pred)
    
    if acc>maxAcc:
        maxAcc=acc
        maxRS=i
        print('Accuracy', maxAcc, 'random_state', maxRS)
```

```python
parameters={'alpha':[.0001,.001,.01,.1,1,10],'random_state':list(range(0,10)), 'selection':['cyclic', 'random'],'normalize':['True','False']}

ls=Lasso()
gcv=GridSearchCV(ls,parameters)
gcv.fit(x_train,y_train)

print(gcv.best_params_)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20,random_state=297)

ls_best=Lasso(alpha=0.1,random_state=6,normalize=True,selection='random')
ls_best.fit(x_train,y_train)
ls_best.score(x_train,y_train)

ls_pred=ls_best.predict(x_test)

ls_score=r2_score(y_test,ls_pred)
ls_score
```
