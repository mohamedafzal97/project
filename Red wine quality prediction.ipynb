# Data Preprocessing

```python
#separating the independent variables and target column into x and y
x=df.drop('quality',axis=1)
y=df['quality']
```

```python
x.sample()
```

```python
y.sample()
```

# removing skewness

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
```

```python
#checking skewness after removing skewness
pd.DataFrame(x).skew()
```

# scaling independent data

using StandardScaler

```python
#mean=0, std=1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

# splitting data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=3)
```

```python
#checking the shape
x_train.shape, y_train.shape, x_test.shape , y_test.shape
```

```python
#importing required libraries
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

# Model building

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
#creating instance 
LR=LogisticRegression()
```

```python
#traing data
LR.fit(x_train,y_train)
```

```python
#predicting data
LR_pred=LR.predict(x_test)
```

```python
#checking accuracy score
print('Accuracy: ',accuracy_score(y_test,LR_pred))
#print confusion matrics
print(confusion_matrix(y_test,LR_pred))
#printing classification report
print(classification_report(y_test,LR_pred))
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
DTC=DecisionTreeClassifier() #creating instance 
DTC.fit(x_train,y_train)  #traing data
DTC_pred=DTC.predict(x_test) #predicting data
```

```python
#checking accuracy score
print('Accuracy: ',accuracy_score(y_test,DTC_pred))
#print confusion matrics
print(confusion_matrix(y_test,DTC_pred))
#printing classification report
print(classification_report(y_test,DTC_pred))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC=RandomForestClassifier(random_state=45) #creating instance 
RFC.fit(x_train,y_train)  #traing data
RFC_pred=RFC.predict(x_test) #predicting data
```

```python
#checking accuracy score
print('Accuracy: ',accuracy_score(y_test,RFC_pred))
#print confusion matrics
print(confusion_matrix(y_test,RFC_pred))
#printing classification report
print(classification_report(y_test,RFC_pred))
```

# Suppport Vector Classifier(SVC)

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
svc.fit(x_train,y_train)
svc_pred=svc.predict(x_test)
```

```python
#checking accuracy score
print('Accuracy: ',accuracy_score(y_test,svc_pred))
#print confusion matrics
print(confusion_matrix(y_test,svc_pred))
#printing classification report
print(classification_report(y_test,svc_pred))
```

# Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
GBC=GradientBoostingClassifier()
```

```python
GBC.fit(x_train,y_train)
GBC_pred=GBC.predict(x_test)
```

```python
#checking accuracy score
print('Accuracy: ',accuracy_score(y_test,GBC_pred))
#print confusion matrics
print(confusion_matrix(y_test,GBC_pred))
#printing classification report
print(classification_report(y_test,GBC_pred))
```

# cross validation

cross validating for all the models builded above to check overfitting if exist.

```python
from sklearn.model_selection import cross_val_score
```

```python
scr=cross_val_score(LR,x,y,cv=10)
print('cross validation score of LogisticRegression: ',scr.mean())
```

```python
scr=cross_val_score(DTC,x,y,cv=10)
print('cross validation score of DecissionTreeClassifier: ',scr.mean())
```

```python
scr=cross_val_score(RFC,x,y,cv=10)
print('cross validation score of RandomForestClassifier: ',scr.mean())
```

```python
scr=cross_val_score(svc,x,y,cv=10)
print('cross validation score of SuperVectorClassifier: ',scr.mean())
```

```python
scr=cross_val_score(GBC,x,y,cv=10)
print('cross validation score of GradientBoostingClassifier: ',scr.mean())
```

# Hyper parameter tuning

From the above cv scores we can see that RandomForestClassifier is giving highest cv score, so we do Hyperparameter tuning to RandomForestClassifier

```python
RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'max_features':['auto', 'sqrt', 'log2'], 'max_depth':[4,5,6,7,8], 'criterion':['gini', 'entropy']}
```

```python
GCV=GridSearchCV(RandomForestClassifier(),parameters, cv=5, scoring='accuracy')
GCV.fit(x_train, y_train) #fitting the data into model
GCV.best_params_
```

```python
GCV_RFC=GCV.best_estimator_
```

```python
GCV_RFC_pred=GCV.best_estimator_.predict(x_test)
```

```python
#accuracy score
accuracy_score(y_test,GCV_RFC_pred)*100
```

# ROC AUC  curve

```python
from sklearn.metrics import plot_roc_curve
```

```python
plot_roc_curve(GCV_RFC,x_test,y_test)
plt.show()
```
