# Data Preprocessing

# Label encoding

```python
df['Sex'].dtypes
```

as wee can see that Sex columns is object type so we have to do label encoding to covert them into integer type

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
le.fit(df['Sex'])
```

```python
le.classes_
```

```python
df['Sex']=le.transform(df['Sex'])
```

```python
df['Sex'].dtypes
```

```python
df['Sex']
```

male is 2, female is 0 and infant is 1

```python
#splitting data into x and y 
x=df.drop('Rings', axis=1)
y=df['Rings']
```

```python
x.sample()
```

```python
y.sample()
```

```python
x=np.array(x)
y=np.array(y)
```

# removing skewness

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
```

```python
#checking skewness after removing skewness
pd.DataFrame(x).skew()
```

# scaling independent data

using StandardScaler

```python
#mean=0, std=1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

# splitting data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=422)
```

```python
#checking the shape
x_train.shape, y_train.shape, x_test.shape , y_test.shape
```

# Model building and Evaluating

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

#creating instance for RandomForestRegressor
RFR = RandomForestRegressor(n_estimators = 200, random_state = 42)

# training model using training data
RFR.fit(x_train, y_train)

#predicting the data
RFR_pred=RFR.predict(x_test)
```

```python
# calculating the mean absolute error
abserror= abs(RFR_pred - y_test)
print('mean absolute error:', np.mean(abserror))
```

```python
#calculating the percentage of mean absolute rror
mean_abs_percentage_error= 100 * (abserror / y_test)
print('mean absolute percentage error:', round(np.mean(mean_abs_percentage_error),2),'%')
```

```python
#calculation of accuracy of model
accuracy = 100 - np.mean(mean_abs_percentage_error)
print('Accuracy:', round(accuracy, 2), '%.')
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression

#creating instance for linearRegression
lm=LinearRegression()

#training the Model
lm.fit(x_train,y_train)


#predicting  the target value
lm_pred=lm.predict(x_test)
```

```python
# calculating the mean absolute error
abserror= abs(lm_pred - y_test)
print('mean absolute error:', np.mean(abserror))
```

```python
#calculating the percentage of mean absolute rror
mean_abs_percentage_error= 100 * (abserror / y_test)
print('mean absolute percentage error:', round(np.mean(mean_abs_percentage_error),2),'%')
```

```python
#calculation of accuracy of model
accuracy = 100 - np.mean(mean_abs_percentage_error)
print('Accuracy:', round(accuracy, 2), '%.')
```

# Super Vector Regressor

```python
from sklearn.svm import SVR
```

```python
#creating the instance
svr=SVR(kernel='rbf')

#training the data
svr.fit(x_train,y_train)

#predicting the data
svr_pred=svr.predict(x_test)
```

```python
# calculating the mean absolute error
abserror= abs(svr_pred - y_test)
print('mean absolute error:', np.mean(abserror))
```

```python
#calculating the percentage of mean absolute rror
mean_abs_percentage_error= 100 * (abserror / y_test)
print('mean absolute percentage error:', round(np.mean(mean_abs_percentage_error),2),'%')
```

```python
#calculation of accuracy of model
accuracy = 100 - np.mean(mean_abs_percentage_error)
print('Accuracy:', round(accuracy, 2), '%.')
```

# KNeighbors Regressor

```python
from sklearn.neighbors import KNeighborsRegressor

#creating the instance
kn = KNeighborsRegressor(n_neighbors=5)

#training the data
kn.fit(x_train,y_train)

#predicting the data
kn_pred=kn.predict(x_test)
```

```python
# calculating the mean absolute error
abserror= abs(kn_pred - y_test)
print('mean absolute error:', np.mean(abserror))
```

```python
#calculating the percentage of mean absolute rror
mean_abs_percentage_error= 100 * (abserror / y_test)
print('mean absolute percentage error:', round(np.mean(mean_abs_percentage_error),2),'%')
```

```python
#calculation of accuracy of model
accuracy = 100 - np.mean(mean_abs_percentage_error)
print('Accuracy:', round(accuracy, 2), '%.')
```

# cross validation

cross validating for all the models builded above to check overfitting if exist.

```python
from sklearn.model_selection import cross_val_score
```

```python
#Random Forest Classifier
scr=cross_val_score(RFR,x,y,cv=10)
print('cross validation score of RandomForestClassifier: ',scr.mean())
```

```python
#Linear Regression
scr=cross_val_score(lm,x,y,cv=10)
print('cross validation score of LinearRegression : ',scr.mean())
```

```python
#SuperVectorRegressor
scr=cross_val_score(svr,x,y,cv=10)
print('cross validation score of SuperVectorRegressor : ',scr.mean())
```

```python
#KNeighborRegressor
scr=cross_val_score(kn,x,y,cv=10)
print('cross validation score of KNeighborRegressor : ',scr.mean())
```

from the above models SuperVectorRegressor is performing better so we continue with it further.
